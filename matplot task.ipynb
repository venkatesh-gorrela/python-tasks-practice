```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting correlation matrix of numerical features
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Plotting distribution of age
sns.histplot(data['Age'].dropna(), bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```

```python
# Plotting count of passengers by gender
sns.countplot(x='Sex', data=data)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.gcf().set_size_inches(3, 5)
plt.title('Count of Passengers by Gender')
plt.show()
```

```python
# Plotting count of passengers by cabin class
sns.countplot(x='Pclass', data=data)
plt.xlabel('Cabin Class')
plt.ylabel('Count')
plt.gcf().set_size_inches(3, 5)
plt.title('Count of Passengers by Cabin Class')
plt.show()
```

```python
# Plotting passenger rate by embarkation point
sns.barplot(x='Embarked', y='PassengerId',data=data,color='#753159')
plt.xlabel('Embarked')
plt.ylabel('PassengerId')
plt.gcf().set_size_inches(3, 5)
plt.title('PassengerId Rate by Embarked')
plt.show()
```
